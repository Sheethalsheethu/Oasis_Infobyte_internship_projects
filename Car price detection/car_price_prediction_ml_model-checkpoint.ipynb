```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
car = pd.read_csv("car data.csv")
car.head()
```

```python
car_new=car.drop(['Car_Name','Selling_type','Owner'],axis='columns')
car_new
```

```python
target=car.Selling_Price
target
```

```python
from sklearn.preprocessing import LabelEncoder
Numerics=LabelEncoder()
```

```python
car_new['Fuel_Type_new']=Numerics.fit_transform(car_new['Fuel_Type'])
car_new['Transmission_new']=Numerics.fit_transform(car_new['Transmission'])
```

```python
car_new
```

```python
car_s_new=car_new.drop(['Fuel_Type','Transmission','Selling_Price'],axis='columns')
car_s_new
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(car_s_new,target)
```

```python
pred = model.predict([[2013,9.54,43000,1,1]])
print(pred)
```
